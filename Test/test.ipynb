```python
from g2p_en import G2p
import json
from itertools import combinations, combinations_with_replacement, product

def reStrip(listWord):
    result = []
    for word in listWord:
        data = word.rstrip("012").lower()
        result.append(data)
    return result

def processingList(listData):
    dictIndex = {}
    for idx, value in enumerate(listData):
        if value in dictIndex:
            dictIndex[value].append(idx)
        else:
            dictIndex[value] = [idx]
    return dictIndex

def givenCombo(listIndex):
    listCombo = []
    for i in range(1, len(listIndex) + 1):
        for combo in combinations(listIndex, i):
            listCombo.append(combo)
    return listCombo

def givenWordCombo(listWordInput, limitLen):
    return list(product(listWordInput, repeat=limitLen))
```

```python
with open(r"P:\Program Files\Python313\AI_assistance\Model\Database\mapping.json", encoding= "utf-8") as f:
    data = json.load(f)

print(g2p("Lionel Messi"))
tempList = reStrip(list(g2p("Lionel Messi")))
dictIndex = processingList(tempList)
# Xây dựng bảng mapping: Dựa trên cách đọc tiếng Anh phiên âm sang tiếng Việt, ví dụ m eh: m é t, r o: r ô
```

```python
result = []
# Value là list các vị trí
for key, value in dictIndex.items():
    if key in data:
        # Nếu từ đó nằm trong database, duyệt qua các bộ từ có thể thay thế
        # Mỗi từ có thể thay thế thì có 1 bộ tổ hợp có thể thay thế
        listCombo = givenCombo(value)
        # Lấy được list các từ có thể thay thế âm tiết, [a, b]
        listWord = data[key]
        # Duyệt qua lần lượt các bộ combo của vị trí
        for combo in listCombo:
            combo = list(combo)
            lenCombo = len(combo)
            # Lấy các combo có thể đưa vào của các từ
            listWordCombo = givenWordCombo(listWord, lenCombo)
            # Duyệt từng combo từ
            for wordCombo in listWordCombo:
                copyData = tempList.copy()
                # Với từng combo từ đưa vào các combo vị trí
                for i in range(len(wordCombo)):
                    addWord =  wordCombo[i]
                    index = combo[i]
                    copyData[index] = addWord
                result.append(copyData)
                with open(r"P:\Program Files\Python313\AI_assistance\Model\word.txt", "a" ,encoding="utf-8") as f:
                    f.write(f"{copyData}\n")

print(result)
```

```python
from g2p_en import G2p
import json
from itertools import combinations, product


def reStrip(listWord):
    result = []
    for word in listWord:
        if "0" in word or "1" in word or "2" in word:
            data = word.rstrip("012").lower()
            result.append(data)
            result.append(" ")
        else:
            result.append(word.lower())
    return result


def givenCombo(listIndex):
    listCombo = []
    for i in range(1, len(listIndex) + 1):
        for combo in combinations(listIndex, i):
            listCombo.append(combo)
    return listCombo

import unicodedata

def remove_tone(text):
    text = text.lower()
    text = unicodedata.normalize('NFD', text)
    text = ''.join(c for c in text if unicodedata.category(c) != 'Mn')
    text = text.replace('đ', 'd')
    return text

def removeSupportLetter(text):
    # Nếu khi split có các từ đứng 1 mình thuộc supportLetter thì xóa ra khỏi từ
    supportLetter = ["b", "c", "d", "đ", "g", "h", "k", "l", "m", "n", "p", "q", "r", "s", "t", "v", "x"]
    temp = text.split()
    listWord = []
    for i in temp:
        if i not in supportLetter:
            listWord.append(i)
    text = " ".join(listWord)
    return text

with open(r"P:\Program Files\Python313\AI_assistance\Model\common-vietnamese-syllables.txt", encoding="utf-8") as f:
    syllables = set(remove_tone(line.strip()) for line in f)

listSyllables = list(syllables)

g2p = G2p()

# Load mapping
with open(r"P:\Program Files\Python313\AI_assistance\Model\Database\mapping.json", encoding="utf-8") as f:
    data = json.load(f)

# G2P
phoneme_list = list(g2p("Mc Donnald"))

# Remove stress number
tempList = reStrip(phoneme_list)

print("Phoneme:", tempList)

# Tìm các vị trí có thể thay
replace_positions = []

for i, p in enumerate(tempList):
    if p in data:
        replace_positions.append(i)

# Sinh tổ hợp vị trí
position_combos = givenCombo(replace_positions)

result = set()

with open(r"P:\Program Files\Python313\AI_assistance\Model\word.txt", "w", encoding="utf-8") as f:

    for combo in position_combos:

        # lấy danh sách lựa chọn cho từng vị trí
        choices = []

        for pos in combo:
            phoneme = tempList[pos]
            choices.append(data[phoneme])

        # sinh tổ hợp từ
        for wordCombo in product(*choices):
            # Báo hiệu cho rằng từ sau khi ghép nối thực sự có nghĩa
            flag = False
            copyData = tempList.copy()

            for i in range(len(combo)):
                index = combo[i]
                copyData[index] = wordCombo[i]

            # Gộp list các từ sau quy đổi âm tiết thành 1 khối
            dataName = removeSupportLetter(remove_tone("".join(copyData)))
            # Lấy các từ sau khi gộp lại coi từ được gộp có thực sự có nghĩa không
            listCheckVN = dataName.split(" ")
            flag = all(x not in listSyllables for x in listCheckVN)
            if not flag:
                if dataName not in result:
                    f.write(f"{dataName}\n")
                result.add(dataName)


print("Total:", len(list(result)))
```
