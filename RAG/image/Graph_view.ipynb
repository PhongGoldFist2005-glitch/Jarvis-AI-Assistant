# RAG Graph Visualization

## Import Thu Vien


```python
# Import thu vien ve graph
import networkx as nx
import matplotlib.pyplot as plt
```

## Tao Va Ket Noi Cac Nut/Canh


```python
# Khoi tao graph co huong
G = nx.DiGraph()

# Danh sach node theo luong xu ly
nodes = [
    "Preprocessing_text",
    "Adding_cache",
    "Planner",
    "Need_previous",
    "Retrieval",
    "Retrieval_STM",
    "Retrieval_LTM",
    "Need_research",
    "Research",
    "Mark_retrieval_stm_done",
    "Mark_retrieval_ltm_done",
    "Mark_research_done",
    "Sync_remove_dup",
    "Remove_duplicate",
    "Enough",
    "LLM_gereration",
    "Update_STM",
    "Need_update_LTM",
    "Update_LTM",
    "Refresh_state",
    "END",
]

# Them node
G.add_nodes_from(nodes)

# Cac canh ket noi theo flow chinh
edges = [
    ("Preprocessing_text", "Adding_cache"),
    ("Adding_cache", "Planner"),
    ("Planner", "Need_previous"),
    ("Planner", "Need_research"),
    ("Planner", "Enough"),
    ("Need_previous", "Retrieval"),
    ("Retrieval", "Retrieval_STM"),
    ("Retrieval", "Retrieval_LTM"),
    ("Retrieval_STM", "Mark_retrieval_stm_done"),
    ("Retrieval_LTM", "Mark_retrieval_ltm_done"),
    ("Mark_retrieval_stm_done", "Sync_remove_dup"),
    ("Mark_retrieval_ltm_done", "Sync_remove_dup"),
    ("Need_research", "Research"),
    ("Research", "Mark_research_done"),
    ("Mark_research_done", "Sync_remove_dup"),
    ("Sync_remove_dup", "Remove_duplicate"),
    ("Enough", "LLM_gereration"),
    ("Remove_duplicate", "LLM_gereration"),
    ("LLM_gereration", "Update_STM"),
    ("Update_STM", "Need_update_LTM"),
    ("Need_update_LTM", "Update_LTM"),
    ("Need_update_LTM", "Refresh_state"),
    ("Update_LTM", "Refresh_state"),
    ("Refresh_state", "END"),
]

# Them canh
G.add_edges_from(edges)

print(f"Tong so node: {G.number_of_nodes()}")
print(f"Tong so canh: {G.number_of_edges()}")
```

## Hien Thi Cau Truc Graph


```python
from pathlib import Path

# Co gang dung graphviz de giam chong cheo, neu khong co thi fallback sang layout khac
try:
    from networkx.drawing.nx_agraph import graphviz_layout

    pos = graphviz_layout(G, prog="dot")
except Exception:
    pos = nx.kamada_kawai_layout(G)

# Ve graph voi mau sac va kich thuoc ro rang
plt.figure(figsize=(20, 12))
nx.draw_networkx(
    G,
    pos=pos,
    with_labels=True,
    node_size=1800,
    node_color="#f2c94c",
    edge_color="#333333",
    font_size=7,
    font_weight="bold",
    arrows=True,
    arrowsize=12,
)

# Luu anh PNG ra file
output_path = Path("AI_assistant/RAG/graph_nx.png")
plt.tight_layout()
plt.savefig(output_path, dpi=220)

print(f"Da luu anh graph: {output_path}")
plt.show()
```
